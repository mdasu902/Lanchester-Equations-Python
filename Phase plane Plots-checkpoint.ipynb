```python
import numpy as np
import matplotlib.pyplot as plt


#import the ODE solver from Scipy
from scipy.integrate import odeint


# define our ODE as a function
def penode(z,t):
    # separate vector y into constituant variable
    x,y = z
    # calculate a vector of derivatives
    dydt = [y,x]
    return dydt

# initial value vector
y0 = [0,  0]
y1 = [0, 0]
y2 = [0,0]
# array of x values to seek y values for
tout = np.linspace(0, 8, 201)



# calculate ODE solutions
soln1 = odeint(penode, y0, tout)
soln2 = odeint(penode, y1, tout)
soln3 = odeint(penode, y2, tout)



# plot results
fig = plt.figure()
ax = fig.add_subplot(111)

# plot scaffolding
ax.set_xlim(-0.5,5)
ax.set_ylim(-0.5, 5)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data',0))


# set the aspect to be one-to-one
ax.set_aspect("equal")

# generate the positions of the arrows
# we use meshgid to produce a 2D array of positions
xq,yq = np.meshgrid(np.linspace(-1, 5, 17),np.linspace(-1,5, 17))

# calculate the directions of the arrows
# we can achieve this by putting the x and y positions
# into a list and passing them to our ODE function
uq, vq = penode([xq, yq],151)

# use the quiver function to plot the arrows
# the pivot argument tells Matplotlib to place the
# middle of the arrow at the (x,y) position
# other options are "tail" and "tip"
ax.quiver(xq,yq,uq,vq, pivot="middle", scale = 90,color="green")


# add the fixed point of the ODE
ax.plot([0],[0],marker="o", mfc="blue", mec="black")
#ax.plot([1],[2],marker="o", mfc="red", mec="red")

xsep=np.array([-1, 10])
ysep=xsep
ax.plot(xsep,ysep, linewidth=4,color="hotpink")

ax.legend(labels=['Directional Field Arrows', '(0,0)', 'Separatrix Line'], loc = 'upper right')

# add labels to the axes and a title to the plot
plt.xlabel('B(t)')
plt.ylabel('R(t)')
plt.title('Phase plane plot for Lanchesters Direct Fire Model',fontsize=22)

# display the plot
fig.set_figwidth(15)
fig.set_figheight(10)

plt.savefig("Direct Fire Model.pdf", dpi=288, bbox_inches="tight")
plt.show()
```
