```python
import matplotlib.pyplot as plt
import numpy as np
import random

np.set_printoptions(suppress=True, precision=0)
fig = plt.figure()
ax = fig.add_subplot(111)


simulation = 10000
n = 1000000
R_win = 0
B_win = 0

rho_R = 0.01042511998  # Value for rho_R
rho_B = 0.05887278716  # Value for rho_B

# Initialize arrays to store outcomes of each simulation
final_R = np.zeros(simulation)
final_B = np.zeros(simulation)

# Loop over simulations
for sim in range(simulation):
    R = np.zeros(n)
    R[0] = 51092
    B = np.zeros(n)
    B[0] = 21500
    
    for i in range(1, n-1):
        if R[i-1] <= 0 or B[i-1] <= 0:
            break
        p = (rho_B * B[i-1]) / (rho_B * B[i-1] + rho_R * R[i-1])
        x = random.uniform(0, 1)
        
        if x <= p:
            R[i] = R[i-1] - 1
            if i < n:
                B[i] = B[i-1]
        else:
            R[i] = R[i-1]
            if i < n:
                B[i] = B[i-1] - 1
                
    # Store final number of soldiers for each simulation
    final_R[sim] = R[i-1]
    final_B[sim] = B[i-1]
    
    if R[i-1] > B[i-1]:
        R_win += 1
    elif B[i-1] > R[i-1]:
        B_win += 1

# Calculate the difference between final populations
D = final_R - final_B
print(len(D))

# Plot histogram
plt.hist(D, bins= np.arange(min(D),max(D)+1),color='b') #edgecolor = 'black')
#print(D)
print(min(D))
print(max(D))
# Plot properties
plt.xlabel('Difference (R_f - B_f)')
plt.ylabel('Frequency')
plt.title('Frequency of Each Army Winning')
plt.grid(True)

# Show plot
fig.set_figwidth(15)
fig.set_figheight(10)
plt.show()

print("America Victories:", R_win)
print("Japanese Victories:", B_win)
```
